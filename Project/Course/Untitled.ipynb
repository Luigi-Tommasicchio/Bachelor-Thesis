## 1. Introduzione ai Modelli di Serie Temporali

Iniziamo con una breve descrizione dei modelli che andrai ad analizzare:
- AR (AutoRegressive): Il modello AR utilizza i valori passati della stessa variabile per prevedere il futuro. È definito dal parametro p che indica il numero di ritardi inclusi nel modello.
- MA (Moving Average): A differenza dell'AR, il modello MA usa gli errori di previsione passati per fare previsioni future. Il parametro q rappresenta il numero di termini di errore inclusi.
- ARMA (AutoRegressive Moving Average): Questo modello combina AR e MA, includendo sia i ritardi della variabile sia gli errori di previsione passati. È definito da due parametri, p e q.
- ARIMA (AutoRegressive Integrated Moving Average): ARIMA estende ARMA includendo la differenziazione (d) per rendere la serie stazionaria, essenziale per trattare trend e stagionalità.
